```python
from ultralytics import YOLO

if __name__ == '__main__':
    model_path = r'F:\实验\蝴蝶兰实验\训练1（batch=64）\weights\best.pt'

    test_img_path = r'F:\实验\蝴蝶兰实验\test2'

    model = YOLO(model_path)

    results = model.predict(test_img_path, project='runs/segment/detect', save=True, conf=0.5)
```

```python
from collections import Counter

cls = []    #图像识别类别
all_cls_dict = {}

if results:
    for i in range(len(results)):
        result = results[i]
        classes_name_dict = result.names   #获取类别名称
        cls.append([int(x) for x in result.boxes.cls.tolist()])   #把每个图像中的识别结果填充到cls中
        img_cls_counter = Counter(cls[i]).items()
        img_cls_counter = sorted(img_cls_counter, key=lambda x: x[0])
        print(f"-----------第{i + 1}张图片检测结果---------------")
        for classes, count in enumerate(img_cls_counter):
            all_cls_dict.update({
                '识别类别': classes_name_dict[classes],
                '检测数量': count[1]
            })
            print(all_cls_dict)
```
